```python
#correlaction table with masking
corr = app_train.corr()
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
f, ax = plt.subplots(figsize=(11, 9))
cmap = sns.diverging_palette(220, 10, as_cmap=True)
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
plt.show()
```

```python
#Perform logisitc regression
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split


#define x and y
X = app_train.drop(['TARGET'], axis=1)
y = app_train['TARGET']


```

```python
#include only float and int columns
X = X.select_dtypes(include=['float64', 'int64'])
X
#remove nan values
X = X.fillna(X.mean())
X.isnull().sum()
```

```python
#split data into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

#fit model
logreg = LogisticRegression()

#fit model
logreg.fit(X_train, y_train)

```

```python
#predict
y_pred = logreg.predict(X_test)

#check accuracy
from sklearn import metrics
print(metrics.accuracy_score(y_test, y_pred))
```

```python
#confusion matrix
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_test, y_pred)
print(confusion_matrix)

#accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, y_pred))
```

```python
#Tensforflow keras for ann
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.layers import Dense, Dropout, Activation, Flatten, Conv2D, MaxPooling2D
from tensorflow.keras.models import Sequential
from tensorflow.keras.callbacks import TensorBoard
import time

#ANN
model = Sequential()

#What architecture?
model.add(Dense(64, input_dim=105, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

#Compile
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

#Fit
model.fit(X_train, y_train, epochs=10, batch_size=32, validation_split=0.2)
```

```python
#Plot Accuracy and Loss
import matplotlib.pyplot as plt
plt.plot(model.history.history['loss'])
plt.plot(model.history.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
```

```python
#plot  val_accuracy
plt.plot(model.history.history['val_accuracy'])
plt.plot(model.history.history['accuracy'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Test', 'Train'], loc='upper left')
plt.show()
```

```python
#best accuracy
print(max(model.history.history['val_accuracy']))
```
